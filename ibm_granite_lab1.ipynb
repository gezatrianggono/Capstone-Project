```python
import pandas as pd
from langchain_community.llms import Replicate
import os

# Setup (sudah ada di colab Anda)
api_token = "your_api_token_here"  # ganti dengan token Anda
os.environ["REPLICATE_API_TOKEN"] = api_token

model = "ibm-granite/granite-3.3-8b-instruct"
output = Replicate(
    model=model,
    replicate_api_token=api_token,
)

# Load data
df = pd.read_csv("BMW sales data (2010-2024) (1).csv")
df_2024 = df[df["Year"] == 2024]
df_2024_sample = df_2024.head(500)

# ========================================
# SOLUSI 1: AGREGASI DATA TERLEBIH DAHULU
# ========================================

# Hitung statistik ringkas dari data
summary_stats = {
    "total_rows": len(df_2024_sample),
    "total_sales_volume": df_2024_sample['Sales_Volume'].sum(),
    "avg_price": df_2024_sample['Price_USD'].mean(),
    "avg_mileage": df_2024_sample['Mileage_KM'].mean(),

    # Top models
    "top_models": df_2024_sample.groupby('Model')['Sales_Volume'].sum().nlargest(5).to_dict(),

    # Regional performance
    "region_sales": df_2024_sample.groupby('Region')['Sales_Volume'].sum().to_dict(),

    # Fuel type distribution
    "fuel_type_sales": df_2024_sample.groupby('Fuel_Type')['Sales_Volume'].sum().to_dict(),

    # Transmission preference
    "transmission_sales": df_2024_sample.groupby('Transmission')['Sales_Volume'].sum().to_dict(),

    # Color preferences
    "color_sales": df_2024_sample.groupby('Color')['Sales_Volume'].sum().nlargest(5).to_dict(),

    # Sales classification
    "sales_class_dist": df_2024_sample['Sales_Classification'].value_counts().to_dict(),

    # Price by engine size correlation
    "engine_price_corr": df_2024_sample.groupby('Engine_Size_L').agg({
        'Price_USD': 'mean',
        'Sales_Volume': 'sum'
    }).to_dict(),

    # High performing configurations
    "high_volume_models": df_2024_sample[df_2024_sample['Sales_Classification'] == 'High'].groupby(
        ['Model', 'Fuel_Type', 'Region']
    )['Sales_Volume'].sum().nlargest(10).to_dict()
}

# Prompt yang lebih efisien dengan data teragregasi
optimized_prompt = f"""
You are an expert data analyst with 50 years of experience in automotive market intelligence.

### Context:
You are analyzing BMW sales data from 2024 (sample of 500 rows from the full dataset spanning 2010-2024).

### Aggregated Data Summary:

**Overall Metrics:**
- Total records analyzed: {summary_stats['total_rows']}
- Total sales volume: {summary_stats['total_sales_volume']:,} units
- Average price: ${summary_stats['avg_price']:,.2f}
- Average mileage: {summary_stats['avg_mileage']:,.0f} KM

**Top 5 Models by Sales Volume:**
{chr(10).join([f"- {model}: {volume:,} units" for model, volume in summary_stats['top_models'].items()])}

**Regional Performance (Total Sales Volume):**
{chr(10).join([f"- {region}: {volume:,} units" for region, volume in summary_stats['region_sales'].items()])}

**Fuel Type Distribution:**
{chr(10).join([f"- {fuel}: {volume:,} units" for fuel, volume in summary_stats['fuel_type_sales'].items()])}

**Transmission Preference:**
{chr(10).join([f"- {trans}: {volume:,} units" for trans, volume in summary_stats['transmission_sales'].items()])}

**Top Colors:**
{chr(10).join([f"- {color}: {volume:,} units" for color, volume in summary_stats['color_sales'].items()])}

**Sales Classification:**
{chr(10).join([f"- {cls}: {count} models" for cls, count in summary_stats['sales_class_dist'].items()])}

**Top 10 High-Volume Configurations (Model-Fuel-Region):**
{chr(10).join([f"- {config}: {volume:,} units" for config, volume in summary_stats['high_volume_models'].items()])}

### Task:
Provide a professional, executive-level analysis in 800-1000 words covering:

1. **Sales Performance Overview**: Interpret the overall 2024 performance with quantitative context
2. **Top Performing Segments**: Analyze which models, regions, and configurations dominate
3. **Market Trends**: Identify emerging patterns in fuel types, regions, and consumer preferences
4. **Price & Engine Insights**: Correlate pricing strategy with sales performance
5. **Strategic Recommendations**: Provide 3-5 actionable insights for BMW's future strategy

### Requirements:
- Write in clear, structured paragraphs (no bullet points in analysis)
- Include both qualitative insights and quantitative data
- Maintain professional business report tone
- Focus on interpretation, not just restating numbers
- Conclude with strategic implications

Generate the complete analysis now.
"""

# Invoke dengan parameter yang lebih besar
parameters = {
    "max_tokens": 4000,  # Naikkan karena input sudah diperkecil
    "top_p": 0.9,
    "top_k": 50,
    "temperature": 0.7,
    "repetition_penalty": 1.1
}

print("=" * 80)
print("OPTIMIZED ANALYSIS WITH AGGREGATED DATA")
print("=" * 80)

response = output.invoke(optimized_prompt, **parameters)
print(response)

print("\n" + "=" * 80)
print("ANALYSIS COMPLETE")
print("=" * 80)
```

```python
# ========================================
# LAB 2 - IBM GRANITE (FIXED VERSION)
# ========================================

# Agregasi data terlebih dahulu (TAMBAHKAN INI SEBELUM LAB 2)
summary_stats = {
    "total_rows": len(df_2024_sample),
    "total_sales_volume": df_2024_sample['Sales_Volume'].sum(),
    "avg_price": df_2024_sample['Price_USD'].mean(),
    "avg_mileage": df_2024_sample['Mileage_KM'].mean(),

    # Top models
    "top_models": df_2024_sample.groupby('Model')['Sales_Volume'].sum().nlargest(5).to_dict(),

    # Regional performance
    "region_sales": df_2024_sample.groupby('Region')['Sales_Volume'].sum().to_dict(),

    # Fuel type distribution
    "fuel_type_sales": df_2024_sample.groupby('Fuel_Type')['Sales_Volume'].sum().to_dict(),

    # Transmission preference
    "transmission_sales": df_2024_sample.groupby('Transmission')['Sales_Volume'].sum().to_dict(),

    # Color preferences
    "color_sales": df_2024_sample.groupby('Color')['Sales_Volume'].sum().nlargest(5).to_dict(),

    # Sales classification
    "sales_class_dist": df_2024_sample['Sales_Classification'].value_counts().to_dict(),

    # High volume configurations
    "high_volume_configs": df_2024_sample[df_2024_sample['Sales_Classification'] == 'High'].groupby(
        ['Model', 'Fuel_Type', 'Region']
    )['Sales_Volume'].sum().nlargest(10).to_dict()
}

print("✅ Data aggregation complete for Lab 2")
print("="*80)

# ========================================
# LAB 2: PARAMETERS (UPDATED)
# ========================================
parameters = {
    "top_k": 40,
    "top_p": 0.92,
    "max_tokens": 4000,  # ✅ Dinaikkan dari 3000
    "min_tokens": 0,
    "random_seed": None,
    "repetition_penalty": 1.15,  # ✅ Ditingkatkan untuk avoid repetisi
    "temperature": 0.75
}

# ========================================
# LAB 2: REFINED PROMPT (DENGAN DATA TERAGREGASI)
# ========================================
refined_focus_prompt = f"""
You are a world-class data analyst and strategic advisor with over 50 years of experience using IBM Granite for advanced business intelligence.

### Context:
You are analyzing BMW's 2024 sales data from a comprehensive dataset spanning 2010-2024. This analysis focuses on the first 500 rows from 2024 to derive actionable insights.

### Aggregated Data Summary:

**Overall Performance Metrics:**
- Total records analyzed: {summary_stats['total_rows']}
- Total sales volume: {summary_stats['total_sales_volume']:,} units
- Average price: ${summary_stats['avg_price']:,.2f}
- Average mileage: {summary_stats['avg_mileage']:,.0f} KM

**Top 5 Models by Sales Volume:**
{chr(10).join([f"  • {model}: {volume:,} units" for model, volume in summary_stats['top_models'].items()])}

**Regional Sales Distribution:**
{chr(10).join([f"  • {region}: {volume:,} units" for region, volume in sorted(summary_stats['region_sales'].items(), key=lambda x: x[1], reverse=True)])}

**Fuel Type Performance:**
{chr(10).join([f"  • {fuel}: {volume:,} units" for fuel, volume in sorted(summary_stats['fuel_type_sales'].items(), key=lambda x: x[1], reverse=True)])}

**Transmission Preference:**
{chr(10).join([f"  • {trans}: {volume:,} units" for trans, volume in summary_stats['transmission_sales'].items()])}

**Top 5 Colors:**
{chr(10).join([f"  • {color}: {volume:,} units" for color, volume in summary_stats['color_sales'].items()])}

**Sales Classification Breakdown:**
{chr(10).join([f"  • {cls}: {count} models" for cls, count in summary_stats['sales_class_dist'].items()])}

**Top 10 High-Performing Configurations (Model-Fuel-Region):**
{chr(10).join([f"  {i+1}. {str(config).replace("', '", " | ").replace("('", "").replace("')", "")}: {volume:,} units" for i, (config, volume) in enumerate(summary_stats['high_volume_configs'].items())])}

### Task:
Provide a comprehensive executive summary structured into THREE key sections (600-800 words total):

#### 1. KEY POINTS DISCUSSED
Synthesize the most critical insights from the 2024 sales data:
- Overall market performance and trends
- Regional dynamics and variations
- Product category performance (models, fuel types, configurations)
- Notable patterns in consumer preferences
Include specific quantitative evidence throughout.

#### 2. DECISIONS MADE
Based on the data analysis, outline strategic conclusions:
- Which segments demonstrate the strongest market position?
- What emerging trends require immediate attention?
- Which product lines or regions warrant increased investment?
- What market shifts indicate necessary strategic pivots?
Support each decision with data points from the summary above.

#### 3. ACTION ITEMS
Provide 6-8 concrete, actionable recommendations:
- Include specific timelines (Q1 2025, H1 2025, FY 2025, etc.)
- Cover: product strategy, regional expansion, technology adoption, marketing focus
- Ensure each action is directly tied to the data insights
- Prioritize actions by potential impact and feasibility

### Format Requirements:
- Write in clear, professional prose under each section heading
- Use short paragraphs for readability (3-5 sentences each)
- Incorporate quantitative data naturally within the text
- Maintain an executive briefing tone
- Avoid generic statements - be specific and data-driven
- Ensure the complete response without truncation

Generate the full executive summary now.
"""

# ========================================
# INVOKE MODEL
# ========================================
print("\n🔍 RUNNING LAB 2: KEY POINTS, DECISIONS & ACTION ITEMS")
print("="*80 + "\n")

response = output.invoke(refined_focus_prompt, **parameters)

print("📊 GRANITE MODEL RESPONSE - LAB 2 (REFINED):")
print("="*80)
print(response)
print("\n" + "="*80)

# ========================================
# VERIFICATION
# ========================================
word_count = len(response.split())
estimated_tokens = int(word_count * 1.3)

print(f"\n📈 OUTPUT METRICS:")
print(f"{'='*60}")
print(f"Word count: {word_count} words")
print(f"Estimated tokens: {estimated_tokens}")
print(f"Target: 600-800 words")

if word_count >= 600:
    print("\n✅ OUTPUT COMPLETE - Analysis successfully generated!")
else:
    print(f"\n⚠️ Output might be truncated (missing ~{600 - word_count} words)")
    print("💡 Recommendation: Increase max_tokens to 5000-6000")

print("="*60)
```
