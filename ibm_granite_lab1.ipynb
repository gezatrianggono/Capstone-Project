```python
import pandas as pd

# baca file dataset
df = pd.read_csv("BMW sales data (2010-2024) (1).csv")

# cek 5 baris pertama
df.head()
```

```python
# Filter dataset hanya untuk tahun 2024
df_2024 = df[df["Year"] == 2024]

print("Jumlah baris untuk tahun 2024:", len(df_2024))
df_2024.head()
```

```python
# Ambil 100 baris pertama saja
df_2024_sample = df_2024.head(500)

print("Jumlah baris yang akan dianalisis:", len(df_2024_sample))
```

```python
all_data_text = df_2024_sample.apply(
    lambda row: " | ".join([f"{col}: {row[col]}" for col in df_2024_sample.columns]),
    axis=1
).tolist()

print("Contoh data yang akan dikirim ke Granite:")
print(all_data_text[:10])  # tampilkan 10 baris pertama biar gak kepanjangan
```

```python
# Refine the prompt to include reviews
all_data_text = df_2024_sample.apply(
    lambda row: " | ".join([f"{col}: {row[col]}" for col in df_2024_sample.columns]),
    axis=1
).tolist()
prompt = f"""
You are an expert data analyst and prompt engineer with 50 years of experience in interpreting business and market data.
You are tasked with analyzing a structured dataset containing BMW car sales information from 2010–2024.

### Dataset Description:
Each row represents a BMW car model with attributes such as:
- Model name and production Year
- Sales region (e.g., Asia, North America, Europe, etc.)
- Vehicle color, fuel type, and transmission type
- Engine size (in liters)
- Mileage (in kilometers)
- Selling price (in USD)
- Sales volume (number of units sold)
- Sales classification (e.g., High, Medium, Low)

### Task:
From this dataset, focus your analysis on **the year 2024 only**, using **the first 500 rows** as the data sample.
Provide a professional, executive-level summary of insights and data-driven observations within a maximum of **500 word max** and add data kualitatif and kuantitatif on output.

### Your analysis must include:
1. **Sales Overview:** Identify the overall sales performance in 2024, including total sales volume trends and regional highlights.
2. **Top Performing Segments:** Identify the most sold models, regions, or configurations (fuel type, transmission, or color).
3. **Market Trends:** Detect emerging market behaviors, such as demand for specific fuel types or high-performing regions.
4. **Price and Engine Insights:** Explore how engine size and price correlate with sales volume or classification.
5. **Strategic Implications:** Conclude with actionable insights that could guide future BMW sales strategies and product focus.

### Output Format:
Present your summary as a clear, concise, and logically structured business report:
- Use short paragraph sections with clear topic focus.
- Avoid bullet points.
- Do not restate dataset columns — interpret them analytically.
- Limit your response to **maximum 500 word**.

Dataset Sample (first 500 rows from 2024):
{all_data_text}
"""

# Invoke the model with example prompt
response = output.invoke(prompt)
# Print the response
print("Granite Model Response:\n")
print(response)
```

```python
# Define the prompt with refined output prompt
refined_outputformat_prompt = f"""
You are a senior data analyst and prompt engineer profesional and komprehensive with over 50 years of experience using IBM Granite for market and business intelligence.
You are continuing the analysis of the BMW sales dataset (2010–2024), focusing exclusively on the year 2024 and using the first 500 rows of data.

Your task is to produce a **structured and concise analytical summary** that complements the executive-level report from the previous analysis and add data kualitatif and kuantitatif on output.

### Objective:
Transform the dataset below into a clear, insight-oriented summary following the sections below:
1. **headings: Key Points Discussed, Decisions Made, and Action Items. Mention timelines. Include only two concise bullet points under each heading.

### Formatting Requirements:
- Present the analysis in **three short, well-structured paragraphs** corresponding to the three sections above.
- Maintain a formal, business-oriented tone suitable for an internal executive report.
- Focus on **interpretation and implications**, not raw numbers.
- Ensure the entire response is within **500 word** maximum.

Dataset sample (first 500 rows from 2024):
{all_data_text}
"""
response = output.invoke(refined_outputformat_prompt)
print("Granite Model Response for refined output format response:\n")
print(response)
```

```python
# Apply default parameters
parameters = {
"top_k": 10,
"top_p": 0.9,
"max_tokens": 200,
"min_tokens": 0,
"random_seed": None,
"repetition_penalty": 1.0,
"stopping_criteria": "length (500 tokens)",
"stopping_sequence": None
}
# Refined prompt with example included
refined_focus_prompt = f"""
You are a world-class data analyst and prompt engineer with over 50 years of experience using IBM Granite for advanced market and business intelligence.
You are analyzing the *BMW sales dataset (2010–2024)* and focusing exclusively on **the year 2024**, using **the first 500 rows** of data.

Summarize this dataset **BMW sales data (2010-2024) (1).csv** by focusing on key points, decisions, and
action items
{all_data_text}
"""
# Invoke the model with the refined focus prompt
response = output.invoke(refined_focus_prompt,
parameters=parameters)
# Print the response
print("Granite Model Response for refined focus area response:\n")
print(response)
```
